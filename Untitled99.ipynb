#Exploring and compensating for null values

```python
#int(df_merged.isnull().sum().sum())
#Another way
(df_merged.isnull().sum())
```

```python
##تعويض القيم الفارغة عن طريق المنوال وهي القيم الاكثر تكرار
# اكثر رسالة تكرار
most_common_message = df_merged['Message'].mode()[0]
```

```python
# تصنيفها
most_common_category = df_merged.loc[
    df_merged['Message'] == most_common_message, 'Category'
].mode()[0]

mask = df_merged['Message'].isnull() & df_merged['Category'].isnull()
```

```python
df_merged.loc[mask, 'Message'] = most_common_message
df_merged.loc[mask, 'Category'] = most_common_category
```

```python
# طباعة البيانات للتأكد انه لا يوجد اي قيم فارغة في كل الاعمدة
print(df_merged.isnull().sum())
```

```python
df_merged['Message'] = df_merged['Message'].fillna("")
```

#**Decision Tree Classifier**

```python
text_cols = ['text', 'Email', 'Subject', 'Sender', 'Recipient']
numeric_cols = ['Attachments', 'Link Count', 'Word Count', 'HTML Tags Count']

y = df_merged['spam_encoded']
```

```python
df_merged['combined_text'] = df_merged[text_cols].astype(str).agg(' '.join, axis=1)

X_text = df_merged['combined_text']
```

```python
# Split data into training and testing sets
X_train_text, X_test_text, y_train, y_test = train_test_split(X_text, y, test_size=0.2)
```

```python
vectorizer = TfidfVectorizer(max_features=3000,stop_words='english',ngram_range=(1,2))
X_train = vectorizer.fit_transform(X_train_text)
X_test = vectorizer.transform(X_test_text)
```

```python
# Train
dtree = DecisionTreeClassifier(class_weight="balanced")
dtree.fit(X_train, y_train)
```

```python
# Predict
y_pred = dtree.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print(classification_report(y_test, y_pred))
```

#**SVM**

```python
text_cols = ['text', 'Email', 'Subject', 'Sender', 'Recipient']
numeric_cols = ['Attachments', 'Link Count', 'Word Count', 'HTML Tags Count']

y = df_merged['spam_encoded']
#  قمت بدمج الأعمدة النصية مع بعض  وذلك لأن(text, Email, Subject, Sender, Recipient)
# لأن خوارزمية SVM لا تتعامل مع النصوص مباشرة بل تحتاج مصفوفة ارقام واحدة
# عند استخدام TF-IDF يتم اعتبار كل صف على أنه رسالة كاملة
# لذلك من الاحسن ان اقوم بجمع كل المعلومات النصية في عمود واحد حتى يتم التصنيف بالشكل المطلوب
```

```python
df_merged['combined_text'] = df_merged[text_cols].astype(str).agg(' '.join, axis=1)
X_text = df_merged['combined_text']
```

```python
#تقسيم البيانات
X_train, X_test, y_train, y_test = train_test_split(X_text, y, test_size=0.2)
```

```python
#تحويل النص  لأرقام
vectorizer = TfidfVectorizer(stop_words='english')

# تدريب  TF-IDF  على بيانات التدريب فقط
X_train_vec = vectorizer.fit_transform(X_train)

# تحويل بيانات الاختبار بنفس القاموس
X_test_vec = vectorizer.transform(X_test)
```

```python
#تم معالجة مشكلة عدم توازن البيانات من خلال استخدام خاصية class_weight=balanced
# داخل النموذج بحيث يتم إعطاء وزن أكبر للفئة الأقل تمثيل Spam مما يساعد في تحسين قدرة النموذج على اكتشاف الرسائل المزعجة وتقليل التحيز للفئة الأكبر
model = LinearSVC(class_weight="balanced", max_iter=5000)
model.fit(X_train_vec, y_train)
```

```python
y_pred_svm = model.predict(X_test_vec)
```

```python
#نموذج Linear SVM
print("SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
```

```python
cm = confusion_matrix(y_test, y_pred_svm)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Confusion Matrix - SVM Spam Classification")
plt.show()
```

```python
# حفظ النموذج
joblib.dump(model, "svm_model.pkl")

# تنزيله على جهازك
files.download("svm_model.pkl")
```
